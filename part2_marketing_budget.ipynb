```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#讀取資料
df = pd.read_excel('C:\\Users\\user\\Desktop\\Titansoft\\Titansoft_BA_Test.xlsx')
#顯示前五筆
display(df.head())
```

```python
df['遊戲頻率'] = 1
df['最近遊戲日期'] = df['日期']
display(df)
```

```python
#最近一次上線時間
#將玩家依帳號進行歸戶
df_group = df.groupby(['帳號','貨幣']).agg({
    '花費金額': 'sum',
    '遊戲頻率':'sum',
    '最近遊戲日期':'max'
}).reset_index()
display(df_group)
```

```python
pd.set_option('display.float_format', lambda x: '%.f' % x)
df_group.describe()
```

```python
df_vip = df_group[(df_group['花費金額']>3215.00)&(df_group['count']>2.00)&(df_group['日期']>'2023-02-28 00:00:00')]
display(df_vip)
```

```python
df_end = df_group[(df_group['花費金額']<3215.00)&(df_group['count']<2.00)&(df_group['日期']<'2023-02-28 00:00:00')]
display(df_end)
```
